## 3-) Definições

**Imagem digital**

Uma imagem digital é a representação numérica de uma imagem via um conjunto de elementos da foto conhecido como pixels. 

**Matriz de imagem**

A matriz de imagem compreende em colunas (M) e linhas (N) que definem os elementos ou pixels de uma dada imagem.

Sendo que o tamanho de uma imagem pode ser definido como:

$Matriz = M*N* (k \ bits)$

O campo de visão (*field of view (FOV)*, em inglês) é o tamanho exibido da imagem. Entretanto, se o mesmo campo de visão for mantido e o tamanho da matriz for aumentado, o tamanho dos pixels serão menores e portanto a resolução será melhorada, em outras palavras teremos uma maior densidade de pixels por $cm^2$, caracterizando um aumento de resolução. 

**Pixel**

O pixel é um elemento que compõe a matriz de imagem, cada pixel é um valor correspondente que representa o nível de brilho da imagem. O tamanho do pixel pode ser determinado como:

$Pixel\ size = FOV / matrix$.

Uma redução do FOV significa um tamanho de pixel menor, resultando em uma melhoria da resolução espacial.

**Bit depth (k-bit)**

O número de bits por pixel é definido como *k bit.* A quantidade de tons ou níveis de cinza de um pixel de uma imagem é igual a $2^{k-bit}$.

- $k-bit\ de \ 2 = 2^2 = 4 \ \text{níveis de cinza}$
- $k-bit \ de \ 8 = 2^8 = 256 \ \text{níveis de cinza}$

Portanto, quanto maior o *bit depth*, maior será a quantidade de tons possíveis. A quantidade de tons possíveis de acordo com o *bit depth,* no caso de tons de cinza, é conhecido como escala de cinza (*grey scale*, em inglês)

**Quantos tons temos na imagem?**

Temos 256 tons de cinza, lembrando que temos 8 bits por pixel, logo $2^8 = 256$

```python
# 12
gradients = {}

for bit in range(1, 11):
  # Criando uma matriz de n bits de profundidade
  # Aqui invertemos o array para ficar uma escala de branco para preto!
  gradient = np.linspace(0, 1, 2**bit)[::-1]
  # Truque para o plot e todas as figuras ficarem de mesma dimensão
  # comentar essa linha e ver o resultado
  gradient = np.vstack([gradient for _ in range(2**bit)])
  
  gradients[f"{bit}"] = gradient.T

# Vamos utilizar o matplotlib para mostrar os gradientes de cinza de acordo com o bit-depth!
f, ax = plt.subplots(2, 5, figsize=(17, 10))

row = 0
col = 0
for i, (key, gradient) in enumerate(gradients.items()):
  if i == 5:
    row = 1
    col = 0
  ax[row][col].imshow(gradient, cmap='gray')
  # Ajustando o título do subplot
  ax[row][col].title.set_text(f'{key}-bit - {2**(int(key))} tons')
  # Removendo os ticks e texto do eixo x e y
  ax[row][col].get_xaxis().set_visible(False)
  ax[row][col].get_yaxis().set_visible(False)
  col +=1
```

**Ajuste de Brilho e contraste**

O brilho de uma imagem é intuitivo, fotos no tiradas no sol são claras, enquanto fotos de noite são em geral escuras. Portanto o brilho de uma imagem refere-se a intensidade dos pixels. Quanto maior o brilho, mais iluminada a imagem aparenta. Um baixo brilho, irá resultar em um aspecto opaco.

O contraste, em contrapartida, refere-se a intensidade relativa entre cores claras e escuras. Ou seja, se olharmos a diferença entre dois pixels, se a intensidade entre eles forem parecidas há um baixo contraste, no entanto se a intensidade for muito diferente, conseguimos diferenciar os dois pixels com facilidade (um alto contraste).


```python
# 20
def adjust_bright_and_contrast(input_img, brightness = 0, contrast = 0):
    buf_img = input_img.copy()
    if brightness != 0:
        if brightness > 0:
            shadow = brightness
            highlight = 255
        else:
            shadow = 0
            highlight = 255 + brightness
        alpha_bright = (highlight - shadow)/255
        gamma_bright = shadow
        
        buf_img = cv2.addWeighted(buf_img, alpha_bright, buf_img, 0, gamma_bright)

    if contrast != 0:
        f = 131*(contrast + 127)/(127*(131-contrast))
        alpha_contrast = f
        gamma_contrast = 127*(1-f)
        
        buf_img = cv2.addWeighted(buf_img, alpha_contrast, buf_img, 0, gamma_contrast)

    return buf_img


f, ax = plt.subplots(2, 3, figsize=(10,7))
ax[0][0].imshow(lenna_img_rgb)
ax[0][0].title.set_text('Original')

ax[0][1].imshow(
    adjust_bright_and_contrast(lenna_img_rgb, -50,0)
)
ax[0][1].title.set_text('Redução Brilho')

ax[0][2].imshow(
    adjust_bright_and_contrast(lenna_img_rgb, 50, 0)
)
ax[0][2].title.set_text('Aumento Brilho')

ax[1][0].imshow(lenna_img_rgb)
ax[1][0].title.set_text('Original')

ax[1][1].imshow(
    adjust_bright_and_contrast(lenna_img_rgb, 0, -50)
)
ax[1][1].title.set_text('Redução Contraste')

ax[1][2].imshow(
    adjust_bright_and_contrast(lenna_img_rgb, 0, 50)
)
ax[1][2].title.set_text('Aumento Contraste')
```

**Limiarização global**

A limiriarização global, nessa técnica aplicamos um limiar (threshold) para cada pixel. Caso o valor seja menor do que o limiar tornamos o valor 0, caso contrário utilizaremos o valor máximo (255, por exemplo).

**Equalização global e local do histograma**  
Há diversas técnicas de processamento de imagens, dentre elas existe a equalização do histograma que permite o ajuste global do contraste da imagem, modificando a imagem de acordo com a distribuição da intensidade do pixel de acordo com o histograma da imagem. Esse processo permite que áreas de baixo contraste ganhem um maior contraste no resultado final.

![https://docs.opencv.org/4.x/histogram_equalization.png](https://docs.opencv.org/4.x/histogram_equalization.png)

```python
# 24
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import requests
wiki_url = "https://i.stack.imgur.com/04OgX.jpg"

img_data = requests.get(url=wiki_url).content
with open("wiki.png", 'wb') as handler:
    handler.write(img_data)
```

```python
# 25
wiki_img = cv2.cvtColor(cv2.imread("wiki.png"), cv2.COLOR_BGR2GRAY)
f, ax = plt.subplots(1, 2, figsize=(10, 4))

# mostrando a imagem original
ax[0].imshow(cv2.cvtColor(wiki_img, cv2.COLOR_BGR2RGB))
# Criando o histograma
ax[1].hist(wiki_img.flatten(),256,[0,256], color = 'r');
```

```python
# 26
# Aplicação da equalização global do histograma utilizando o openCV
wiki_img_equ = cv2.equalizeHist(wiki_img)

# Imprimindo a imagem
f, ax = plt.subplots(2,2, figsize=(10,8))
# Imagem original
ax[0][0].imshow(cv2.cvtColor(wiki_img, cv2.COLOR_BGR2RGB))
# Histograma da imagem original
ax[0][1].hist(wiki_img.flatten(),256,[0,256], color = 'r');
# Imagem equalizada
ax[1][0].imshow(cv2.cvtColor(wiki_img_equ, cv2.COLOR_BGR2RGB))
# Histograma da imagem equalizada
ax[1][1].hist(wiki_img_equ.flatten(),256,[0,256], color = 'r');
```

A equalização global do histograma é bom quando o histograma da imagem está restrita em uma região particular, como no caso acima. Em situações em que a imagem apresenta uma grande variação de intensidade de pixels, ou seja o histograma apresenta uma boa cobertura (com regiões claras e escuras), a equalização global do histograma costuma falhar.

Para aumentarmos o contraste em imagens em que há uma ampla distribuição de intensidade de pixels, podemos utilizar uma equalização local do histograma. Há diversos métodos para realizar a equalização local, sendo que um deles é conhecido como CLAHE (do inglês, *Contrast Limited Adaptive Histogram Equalization*).

```python
# 27
import requests
tsukuba = "https://static-02.hindawi.com/articles/isrn/volume-2012/163285/figures/163285.fig.003a.jpg"

img_data = requests.get(url=tsukuba).content
with open("tsukuba.png", 'wb') as handler:
    handler.write(img_data)
```

```python
# 28
tsukuba_img = cv2.cvtColor(cv2.imread("tsukuba.png"), cv2.COLOR_BGR2GRAY)
tsukuba_img_equ = cv.equalizeHist(tsukuba_img)
f, ax = plt.subplots(2,2, figsize=(13,10))
ax[0][0].imshow(cv2.cvtColor(tsukuba_img, cv2.COLOR_BGR2RGB))
ax[0][1].hist(tsukuba_img.flatten(),256,[0,256], color = 'r');
ax[1][0].imshow(cv2.cvtColor(tsukuba_img_equ, cv2.COLOR_BGR2RGB))
ax[1][1].hist(tsukuba_img_equ.flatten(),256,[0,256], color = 'r');
```

```python
# 29

# CLAHE
# Dividimos a imagem em tiles
# Aplicamos uma equalização global em cada tile
# Aplicamos um limitador de constrate
# Remoção de artefatos das bordas de cada tile -> interpolação bilinear

# Criando um objeto CLAHE
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
# Aplicando na imagem de interesse `tsukuba_img`, nesse caso
cl1 = clahe.apply(tsukuba_img)

# Plot

f, ax = plt.subplots(2,3, figsize=(18,9))
# Imagem original
ax[0][0].imshow(cv2.cvtColor(tsukuba_img, cv2.COLOR_BGR2RGB))
ax[0][0].title.set_text('Original')
# Eq. Global
ax[0][1].imshow(cv2.cvtColor(tsukuba_img_equ, cv2.COLOR_BGR2RGB))
ax[0][1].title.set_text('Equalização global')
# Eq. Local
ax[0][2].imshow(cv2.cvtColor(cl1, cv2.COLOR_BGR2RGB))
ax[0][2].title.set_text('Equalização local')

# Histograma original
ax[1][0].hist(tsukuba_img.flatten(),256,[0,256], color = 'r')
# Histograma Eq. Global
ax[1][1].hist(tsukuba_img_equ.flatten(),256,[0,256], color = 'r')
# Histograma Eq. Local
ax[1][2].hist(cl1.flatten(),256,[0,256], color = 'r')
plt.show()
```

*Translação*

Agora, translacionamos a imagem (fazemos um deslocamento), para tal utilizamos uma **matriz de transformação afim** do tipo:

$M = \begin{bmatrix}
1 & 0 & t_x\\
0 & 1 & t_y
\end{bmatrix}$

Sendo que valores negativos para $t_x$ deslocaram a imagem para esquerda, e valores positivos para a direita. E valores negativos de $t_y$ deslocaram a imagem para cima, e valores positivos para baixo

**Cisalhamento**

Por fim, podemos utilizar uma matriz afim para aplicar um cisalhamento a imagem.

Para tal utilizamos uma matriz afim semelhante quando aplicamos alguma translação a imagem.

$M = \begin{bmatrix}
1 & cx & 0\\
cy & 1 & 0
\end{bmatrix}$

```python
# 33
rows, cols, _ = lenna_img_rgb.shape

M_45 = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)
M_90 = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)

lenna_45 = cv2.warpAffine(lenna_img_rgb, M_45, (rows, cols))
lenna_90 = cv2.warpAffine(lenna_img_rgb, M_90, (rows, cols))

f, ax = plt.subplots(1, 3, figsize=(10, 5))

ax[0].imshow(lenna_img_rgb)
ax[0].title.set_text('Original')

ax[1].imshow(lenna_45)
ax[1].title.set_text('Rot 45')

ax[2].imshow(lenna_90)
ax[2].title.set_text('Rot 90')
```
