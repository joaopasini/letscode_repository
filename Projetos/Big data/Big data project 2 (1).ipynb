```
schema_df = spark.read.csv(path='/databricks-datasets/power-plant/data/Sheet1.tsv', header=True, sep='\t', inferSchema=True) #choosing schema as header
df = spark.read.csv(path='/databricks-datasets/power-plant/data/', header=True, sep='\t', schema=schema_df.schema) # reading from multiple files and using previous schema
display(df)
```

```
import pyspark.sql.functions as f

data_agg = df.agg(*[f.count(f.when(f.isnull(c), c)).alias(c) for c in df.columns])
data_agg.show()
```

```
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(inputCols=['AT', 'V', 'AP', 'RH','PE'], outputCol='features')

assembler_df = assembler.transform(df)
assembler_df.show()
```

```
assembler_df.select('features').show()
```

```
df_train, df_test = assembler_df.randomSplit([0.8,0.2])
```

```
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(featuresCol='features', labelCol='PE')
lr_model = lr.fit(df_train)
```

```
predictions = lr_model.transform(df_test)
display(predictions)
```

```
from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(labelCol='PE',
                   predictionCol='prediction', 
                   metricName='mae')

mae = evaluator.evaluate(predictions)
print(mae)
```
