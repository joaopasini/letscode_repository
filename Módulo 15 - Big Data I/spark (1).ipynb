### Machine Learning com Sklearn

```python
from sklearn.model_selection import train_test_split

y = df_pandas['cnt']
X = df_pandas.drop(['cnt', 'instant', 'dteday'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, y_pred)
```

### Machine Learning com Spark

```python
from pyspark.sql import HiveContext
contexto = HiveContext(sc)

df = contexto.table("ada857.bikesharing") ### leitura via hive 
```

```python
df = spark.read.load('/databricks-datasets/bikeSharing/data-001/hour.csv', format='csv', sep=',', header=True, inferSchema=True)  ### leitura via csv
```

```python
y = df.select('cnt')
X = df.drop('cnt', 'instant', 'dteday')
```

```python
X.show()
```

```python
df_train, df_test = df.randomSplit([0.8, 0.2])
```

```python
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(inputCols=X.columns, outputCol = 'features')
assembler_df = assembler.transform(df)
```

```python
assembler_df.show()
```

```python
assembler_df.select('features').show()
```

```python
training_data, test_data = assembler_df.randomSplit([0.8, 0.2])
```

```python
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(featuresCol='features', labelCol='cnt')
model = lr.fit(training_data)
```

```python
predictions = model.transform(test_data) ###predict
```

```python
predictions.show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(
    labelCol = 'cnt',
    predictionCol ='prediction',
    metricName = 'mae'
)
```

```python
mae = evaluator.evaluate(predictions)
print(mae)
```

```python
predictions.select('prediction').show()
```

### Etapas de ML no Spark

1. Aquisição dos dados (via csv ou hive por exemplo)
2. Faz a análise exploratória (em caso de muitos dados particiona esses dados para isto trabalhando no pandas)
3. Usa o dataframe do spark (sem o pandas) para treinamento de modelos
4. Cria uma coluna de features no dataframe usando o vectorAssembler 
5. Divide os dados usando a função randomSplit
6. Cria uma instância do modelo de ML desejado passando o nome da coluna de features e do target
7. Realiza as predições na base de teste usando a função transform
8. Cria uma instância para a métrica de avaliação informando as colunas de predição e rotulos (y_test, y_pred)
9. Usa o evaluator para gerar a métrica de avaliação (https://spark.apache.org/docs/latest/mllib-evaluation-metrics.html)
